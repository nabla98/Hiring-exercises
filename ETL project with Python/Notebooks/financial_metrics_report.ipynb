```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
from mysql.connector import Error
```

```python

# 1. Conectar a la base de datos
conexion = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="stock_market_data"
)

# 2. Leer los datos de la vista
query = """
SELECT ticker, avg_daily_volatility
FROM vista_volatilidad_empresas
ORDER BY avg_daily_volatility DESC
LIMIT 5;
"""
df = pd.read_sql(query, conexion)

# 3. Graficar
plt.figure(figsize=(10,6))
plt.bar(df['ticker'], df['avg_daily_volatility'], color='skyblue')
plt.xlabel('Empresa (Ticker)')
plt.ylabel('Volatilidad Promedio Diaria')
plt.title('Top 5 Empresas Más Volátiles - Últimos 6 Meses')
plt.grid(axis='y')
plt.show()

# 4. Cerrar conexión
conexion.close()
```
